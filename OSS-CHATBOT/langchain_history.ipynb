```python
import os
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI

# 1) .env ë¶ˆëŸ¬ì˜¤ê¸°
load_dotenv()

# 2) LangChainìš© LLM ê°ì²´ ë§Œë“¤ê¸° (Azure OpenAI v1 ë°©ì‹)
model = ChatOpenAI(
    model=os.getenv("DEPLOYMENT_NAME"),               
    api_key=os.getenv("OPENAI_API_KEY"),            
    base_url=os.getenv("ENDPOINT_URL").rstrip("/") + "/openai/v1/",
    temperature=0.0,
)
```

```python
# 3) ë©”ì‹œì§€ íˆìŠ¤í† ë¦¬
from langchain_core.chat_history import InMemoryChatMessageHistory
from langchain_core.runnables.history import RunnableWithMessageHistory, BaseChatMessageHistory
from langchain_core.messages import HumanMessage

store = {}
```

```python
def get_session_history(session_id: str) -> BaseChatMessageHistory:

    if session_id not in store:
        store[session_id] = InMemoryChatMessageHistory()

    return store[session_id]
```

```python
with_message_history = RunnableWithMessageHistory(model, get_session_history)
```

```python
config = {"configurable": {"session_id": "abc2"}}

response = with_message_history.invoke(
    [HumanMessage(content="ì•ˆë…•? ë‚˜ëŠ” ê¹€í˜„ìˆ˜ì•¼")],
    config=config,
)

print(response.content)
```
